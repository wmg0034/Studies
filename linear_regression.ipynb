```python
# prompt: Generate a dataset with two continuous variables, x and y. Give them a roughly linear relationship with randomized coefficients and Gaussian noise. Plot the data set as a scatter plot.

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Generate data
# np.random.seed(0)  # for reproducibility
num_points = 100
x = np.random.rand(num_points) * 10
y = np.random.rand(1)*x + np.random.rand(1) + np.random.randn(num_points) * 2  # Linear relationship with noise

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Generated Data', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of Generated Dataset')
plt.grid(True)
plt.legend()
plt.show()
```

# Non-object

```python
def lin_reg(x,y):
  learning_rate   = 0.01
  threshold       = 0.000000001

  m, b = 0, np.mean(y)
  predictions = b * np.ones(len(y))# Initialize to mean
  error = y - predictions
  mse = np.dot( error, error ) / len(error)
  converged = False

  while not converged:
    dldm  = -2 * np.dot( x, error ) / len(error)
    dldb  = -2 * np.sum(error) / len(error)
    m     = m - learning_rate * dldm
    b     = b - learning_rate * dldb

    old_mse = mse
    predictions = m * x + b
    error = y - predictions
    mse = np.dot( error, error ) / len(error)
    print(mse)

    if abs(mse - old_mse) < threshold: #Exit if MSE has not meaningfully changed
      converged = True

  t = np.linspace(min(x), max(x), num=50)
  l = m * t + b

  plt.figure(figsize=(8, 6))
  plt.scatter(x, y, label='Regression plot', alpha=0.7)
  plt.plot(t, l, color='red')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'Linear regression with m: {m}, b: {b}')
  plt.grid(True)
  plt.legend()
  plt.show()


  return
```

```python
lin_reg(x,y)
```

# OOP

```python
class MyLinearRegression():
    def __init__(self, learning_rate, threshold):
        self.learning_rate  = learning_rate
        self.threshold      = threshold

    def get_slope(self):
        return self.slope
    def set_slope(self, value):
        self.slope = value

    def get_intercept(self):
        return self.intercept
    def set_intercept(self, value):
         self.intercept = value

    def fit(self, x, y):
        m, b = 0, np.mean(y)
        predictions = b * np.ones(len(y))# Initialize to mean
        error = y - predictions
        mse = np.dot( error, error ) / len(error)
        converged = False


        while not converged:
            dldm  = -2 * np.dot( x, error ) / len(error)
            dldb  = -2 * np.sum(error) / len(error)
            m     = m - self.learning_rate * dldm
            b     = b - self.learning_rate * dldb

            old_mse = mse
            predictions = m * x + b
            error = y - predictions
            mse = np.dot( error, error ) / len(error)
            print(mse)

            if abs(mse - old_mse) < self.threshold: #Finish if MSE has not meaningfully changed
                converged = True

        self.slope      = m
        self.intercept  = b

        return

    def show_plot(self):
        t = np.linspace(min(x), max(x), num=50)
        l = self.get_slope() * t + self.get_intercept()
        plt.figure(figsize=(8, 6))
        plt.scatter(x, y, label='Regression plot', alpha=0.7)
        plt.plot(t, l, color='red')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(f'Linear regression with m: {self.slope}, b: {self.intercept}')
        plt.grid(True)
        plt.legend()
        plt.show()

        return
```

```python
model = MyLinearRegression(learning_rate = .001, threshold = 10 ** -9)
model.fit(x,y)
```
